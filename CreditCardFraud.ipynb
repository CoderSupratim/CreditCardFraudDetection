### Imbalanced data and unscaled Time and amount column in dataframe

```python
from sklearn.preprocessing import RobustScaler
from sklearn.model_selection import train_test_split
```

```python
# Scaling and reshaping Time and amount
scaler = RobustScaler()
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1,1))
df['Time'] = scaler.fit_transform(df['Time'].values.reshape(-1,1))
```

```python
df_copy2 = df.copy() # creating copy as checkpoint
```

```python
df = df.sample(frac=1)

# Shuffle dataframe rows
new_df = df.sample(frac=1, random_state=42)

new_df.head()
```

Now we can see that the data is balanced

### splitting and Scaling the dataset

```python
train, valid, test = np.split(new_df, [int(0.6*len(new_df)), int(0.8*len(new_df))]) # splitting the dataframe into train, valid and test data
```

```python
# Scaling the dataset
def scale_dataset(dataframe):
  X = dataframe[dataframe.columns[:-1]].values
  Y = dataframe[dataframe.columns[-1]].values

  data = np.hstack((X, np.reshape(Y, (-1, 1))))

  return data, X, Y
```

```python
train, X_train, Y_train = scale_dataset(train)
valid, X_valid, Y_valid = scale_dataset(valid)
test, X_test, Y_test = scale_dataset(test)

X_train = scaler.fit_transform(X_train)
X_valid = scaler.transform(X_valid)
X_test = scaler.transform(X_test)
```

### Model Training

```python
from sklearn.metrics import precision_recall_curve, auc, average_precision_score
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
# from sklearn.neighbors import KNeighborsClassifier
# from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
```

### Training and parameter tuning of each Model

```python
# listing all models as dictionary
models = {
    'LogisticRegression': LogisticRegression(solver='liblinear'),
    # 'KNN': KNeighborsClassifier(),
    # 'SVC': SVC(probability=True),
    'DecisionTree': DecisionTreeClassifier()
}

# Listing possible parameters as dictionary
param_grid = {
    'LogisticRegression': {
        'penalty': ['l1', 'l2'],
        'C': [0.001, 0.01, 0.1, 1, 10, 100]
    },
    # 'KNN': {
    #     'n_neighbors': [3, 5, 7, 9],
    #     'weights': ['uniform', 'distance']
    # },
    # 'SVC': {
    #     'C': [0.1, 1, 10],
    #     'kernel': ['linear', 'rbf']
    #  },

    'DecisionTree': {
        'max_depth': [None, 5, 10, 20],
        'min_samples_split': [2, 5, 10]
    }
}

# choosing best performing models
best_models = {}
for name, model in models.items():
    grid_search = GridSearchCV(model, param_grid[name], cv=5, scoring='average_precision')
    grid_search.fit(X_train, Y_train)
    best_models[name] = grid_search.best_estimator_

# Use the best models for predictions on Validation Data
best_model_predictions = {}
for name, model in best_models.items():
    Y_pred_proba = model.predict_proba(X_valid)[:, 1]  # Get probability estimates for the positive class
    precision, recall, _ = precision_recall_curve(Y_valid, Y_pred_proba)
    auprc = auc(recall, precision)

    # Plot Precision-Recall curve
    plt.figure()
    plt.plot(recall, precision, color='blue', lw=2, label='Precision-Recall curve (AUPRC = %0.2f)' % auprc)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title(f'Precision-Recall Curve for {name}')
    plt.legend(loc="lower right")
    plt.show()

    print(f"{name} AUPRC:", auprc)
```

```python
# Using the best models for predictions on Test Data
best_model_predictions_test = {}
for name, model in best_models.items():
    Y_pred_proba_test = model.predict_proba(X_test)[:, 1]  # Get probability estimates for the positive class
    precision_test, recall_test, _ = precision_recall_curve(Y_test, Y_pred_proba_test)
    auprc_test = auc(recall_test, precision_test)

    # Plot Precision-Recall curve for Test Data
    plt.figure()
    plt.plot(recall_test, precision_test, color='blue', lw=2, label='Precision-Recall curve (AUPRC = %0.2f)' % auprc_test)
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.ylim([0.0, 1.05])
    plt.xlim([0.0, 1.0])
    plt.title(f'Precision-Recall Curve for {name} (Test Data)')
    plt.legend(loc="lower right")
    plt.show()

    print(f"{name} AUPRC on Test Data:", auprc_test)
```
